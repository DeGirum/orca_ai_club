```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Solution 1: Vectors
vector1 = np.array([-10 / np.sqrt(2), 10 / np.sqrt(2)])
vector2 = np.array([15 / np.sqrt(2), 15 / np.sqrt(2)])
resultant = np.array([5 / np.sqrt(2), 25 / np.sqrt(2)])

# Calculate the magnitude of the resultant vector
magnitude = np.linalg.norm(resultant)

# Calculate the angle of the resultant vector in degrees
theta_rad = np.arctan2(resultant[1], resultant[0])
theta_deg = np.degrees(theta_rad)

# Create a square figure and axis
fig, ax = plt.subplots(figsize=(6, 6))

# Plotting vectors
ax.quiver(
    0,
    0,
    vector1[0],
    vector1[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="r",
    label="Vector 1",
)
ax.quiver(
    vector1[0],
    vector1[1],
    vector2[0],
    vector2[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="b",
    label="Vector 2",
)
ax.quiver(
    0,
    0,
    resultant[0],
    resultant[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="g",
    label="Resultant",
)
ax.set_xlim(-20, 20)  # Adjust the x-axis limits to maintain a square grid
ax.set_ylim(-20, 20)  # Adjust the y-axis limits to maintain a square grid
ax.axhline(0, color="k", linewidth=1.5)
ax.axvline(0, color="k", linewidth=1.5)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Vector Addition: Triangle Law")
ax.legend()
ax.grid()

# Make the aspect ratio equal
ax.set_aspect("equal")

# Display the magnitude and angle on the plot
ax.text(
    -19,
    17,
    f"Resultant Magnitude: {magnitude:.2f} mph",
    bbox=dict(facecolor="white", alpha=0.7),
)
ax.text(
    -19,
    15,
    f"Resultant Direction: {theta_deg:.2f}° North of East",
    bbox=dict(facecolor="white", alpha=0.7),
)

plt.show()
```

```python
# Solution 2 - Matrices
R = np.array([[100, 50], [200, 255]])
G = np.array([[150, 100], [100, 255]])
B = np.array([[200, 150], [50, 255]])

# Creating an image from the R, G, B matrices
original_image = np.stack((R, G, B), axis=-1)

# Halving the values in the matrices
R_ = R / 2
G_ = G / 2
B_ = B / 2

# Creating an image from the halved R, G, B matrices
reduced_image = np.stack((R_, G_, B_), axis=-1)

# Plotting the original and brightness-reduced images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(original_image.astype(np.uint8))
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(reduced_image.astype(np.uint8))
axes[1].set_title("Brightness Reduced by 50%")
axes[1].axis("off")

plt.show()
```

```python
# Solution 3: Linear Transformations
original_vector = np.array([3, 4])
rotated_vector = np.array([4, -3])

# Create a square figure and axis
fig, ax = plt.subplots(figsize=(6, 6))

# Plotting vectors before and after rotation
ax.quiver(
    0,
    0,
    original_vector[0],
    original_vector[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="r",
    label="Original Vector",
)
ax.quiver(
    0,
    0,
    rotated_vector[0],
    rotated_vector[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="b",
    label="Rotated Vector",
)
ax.set_xlim(-4, 5)
ax.set_ylim(-4, 5)
ax.axhline(0, color="k", linewidth=1.5)
ax.axvline(0, color="k", linewidth=1.5)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Linear Transformation: Rotation")
ax.legend()
ax.grid()

# Make the aspect ratio equal
ax.set_aspect("equal")

plt.show()
```
