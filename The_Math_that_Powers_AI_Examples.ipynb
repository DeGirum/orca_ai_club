```python
# Simple 2D Vector Plotting example

# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Defining a 2D vector (6, -6)
vector = np.array([6, -6])

# Plotting the vector
plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-1, 7)
plt.ylim(-7, 1)

# Adding coordinate axes
plt.axhline(0, color='k', linewidth=1.5)
plt.axvline(0, color='k', linewidth=1.5)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Vector')
plt.grid()
plt.show()


```

```python
# 2D Vector Addition example
vector1 = np.array([5, 1])
vector2 = np.array([-1, 3])

# Adding the vectors
resultant = vector1 + vector2

# Setting up the figure and axis
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Triangle Law
axes[0].quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
axes[0].quiver(vector1[0], vector1[1], vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector 2')
axes[0].quiver(0, 0, resultant[0], resultant[1], angles='xy', scale_units='xy', scale=1, color='g', label='Resultant')
axes[0].set_title('Triangle Law of Vector Addition')
axes[0].grid(True)
axes[0].legend()

# Add coordinate axes to the first subplot
axes[0].axhline(0, color='k', linewidth=1.5)
axes[0].axvline(0, color='k', linewidth=1.5)

# Parallelogram Law
axes[1].quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='r', alpha=0.5)
axes[1].quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='b', alpha=0.5)
axes[1].quiver(0, 0, resultant[0], resultant[1], angles='xy', scale_units='xy', scale=1, color='g', label='Resultant')
axes[1].quiver(vector2[0], vector2[1], vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='r', alpha=0.5)
axes[1].quiver(vector1[0], vector1[1], vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='b', alpha=0.5)
axes[1].set_title('Parallelogram Law of Vector Addition')
axes[1].grid(True)
axes[1].legend()

# Add coordinate axes to the second subplot
axes[1].axhline(0, color='k', linewidth=1.5)
axes[1].axvline(0, color='k', linewidth=1.5)

# Adjust the plot limits based on the vectors for both axes
for ax in axes:
    ax.set_xlim(-2, max(resultant[0]+2, 10))
    ax.set_ylim(-2, max(resultant[1]+2, 10))
    ax.set_xlabel('X')
    ax.set_ylabel('Y')

plt.tight_layout()
plt.show()
```

```python
# Color Vector example
# Defining color vectors
red = [255, 0, 0]
green = [0, 255, 0]
blue = [0, 0, 255]

# Displaying the colors
plt.imshow([[red, green, blue]])
plt.axis('off')
plt.show()
```

```python
# Color Vector addition
newColor = np.array(red) + np.array(blue)


# Displaying the colors: red, blue, and red+blue
plt.imshow([[red, blue, newColor]])
plt.axis('off')
plt.show()
```

```python
# Scalar - Color multiplication Example
scalar = 0.5

# Scaling the color vectors
scaled_red = np.array(red) * scalar
scaled_green = np.array(green) * scalar
scaled_blue = np.array(blue) * scalar

# Clip the scaled color vectors to valid range [0, 255] and convert to integer
scaled_red = np.clip(scaled_red, 0, 255).astype(int)
scaled_green = np.clip(scaled_green, 0, 255).astype(int)
scaled_blue = np.clip(scaled_blue, 0, 255).astype(int)

# Displaying the scaled colors
plt.imshow([[scaled_red, scaled_green, scaled_blue]])
plt.axis('off')
plt.show()
```

```python
# 3D Vector Addition example
from mpl_toolkits.mplot3d import Axes3D

# Defining two 3D vectors
vector1 = np.array([2, 1, 1])
vector2 = np.array([1, 2, 2])

# Adding the vectors
resultant = vector1 + vector2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plotting the vectors
ax.quiver(0, 0, 0, vector1[0], vector1[1], vector1[2], color='r', label='Vector 1', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, vector2[0], vector2[1], vector2[2], color='b', label='Vector 2', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, resultant[0], resultant[1], resultant[2], color='g', label='Resultant', arrow_length_ratio=0.1)

# Set limits to better visualize the vectors
ax.set_xlim([0, max(vector1[0], vector2[0], resultant[0]) + 1])
ax.set_ylim([0, max(vector1[1], vector2[1], resultant[1]) + 1])
ax.set_zlim([0, max(vector1[2], vector2[2], resultant[2]) + 1])

# Setting labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Vector Addition')
ax.legend()

plt.show()

```

```python
# Using slope of x^2 to find the minimum value of x^2 (gradient descent sneek-peek)

# Define f(x) = x^2 and its derivative
def f(x):
    return x**2

def df(x):  # Derivative (slope) at x
    return 2*x

# Gradient descent parameters
alpha = 0.1  # learning rate
num_iterations = 20
x_value = 10  # starting point

# Values for plotting the function
x = np.linspace(-10, 10, 400)
y = f(x)

# Plotting the function
plt.plot(x, y, 'r', label='f(x) = x^2')

# Gradient Descent in action
for i in range(num_iterations):
    # Marking the current x_value on the curve
    plt.scatter(x_value, f(x_value), color='blue')

    # Plotting the tangent line (derivative visualization)
    tangent_x = np.linspace(x_value - 1, x_value + 1, 2)
    tangent_y = df(x_value)*(tangent_x - x_value) + f(x_value)

    # Plot thicker and solid lines with outlines
    plt.plot(tangent_x, tangent_y, 'g-', linewidth=2, alpha=0.5)  # Thick green line for outlines
    plt.plot(tangent_x, tangent_y, 'k-', linewidth=3, alpha=0.2)  # Original tangent line


    # Gradient descent step
    x_value -= alpha * df(x_value)

# Making the plot visually appealing
plt.title("Gradient Descent on f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
