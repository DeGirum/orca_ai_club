What are the two components we need for a successful game run of Rock-Paper-Scissors?

1. User input
2. Computer input

Let's start by defining the possible choices for the computer and the user. Recall the concept we use to store multiple values in a single variable: lists! 

```python
# Here is a way we can define a list of choices 
list_of_choices = ['rock', 'paper', 'scissors']
```
```

````python
list_of_choices = ["rock", "paper", "scissors"]
````

Now, we are ready to take a random choice from the computer and the user. We can use the `random` module to do this. 
Here is an example of how random.choice() works:

```python
import random
list_of_choices = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
random.choice(list_of_choices)
```
```

random.choice() will return a random element from the list.
Here is a link to the python documentation for the random module: 
https://docs.python.org/3/library/random.html#random.choice

````python
def generate_computer_choice() -> str:
    """Generate a random choice from rock, paper, or scissors"""
    ## Initialize a computer_choice variable to a random choice of rock, paper, or scissors
    computer_choice = random.choice(list_of_choices)
    return computer_choice
````

Great work! We have access to the choice generated by the computer. Now it's time to get an input from the user of this game.
Recall, we can use the `input()` function to get an input from the user. 
```python
user_choice = input("Please enter your choice: ")
```
```

````python
def get_user_choice() -> str:
    """Get the user's choice"""
    user_choice = input("Enter your choice: ")
    ## We want to make sure that the user's choice is valid. Recall the concept of assertions
    user_choice = user_choice.lower()
    assert user_choice in list_of_choices, "Invalid choice"
    return user_choice
````

Excellent! We are now getting to the bedrock of the game, i.e. given two inputs (user choice and computer choice), deciding who wins! 

To refresh your memory, 
1. Rock beats scissors
2. Scissors beats paper
3. Paper beats rock

HINT: To accomplish these rules, we can use the `if-elif-else` construct. 
And in case you didn't think of it, both the user and the computer can have the same choice. Make sure you account for this case as well.

````python
def get_game_result(computer_choice: str, user_choice: str) -> str:
    """Decides the result of the game

    Args:
        computer_choice (str): The computer's choice
        user_choice (str): The user's choice

    Returns:
        str: The result of the game
    """
    ## What are the different cases that we need to consider?
    ### What if the user and computer choose the same thing?
    if computer_choice == user_choice:
        return "Tie"
    elif computer_choice == "rock":
        if user_choice == "paper":
            return "User wins"
        else:
            return "Computer wins"
    elif computer_choice == "paper":
        if user_choice == "scissors":
            return "User wins"
        else:
            return "Computer wins"
    elif computer_choice == "scissors":
        if user_choice == "rock":
            return "User wins"
        else:
            return "Computer wins"
    else:
        return "Invalid choice"
````

Now that we have all the pieces, let's put them together and write the game logic.

````python
def play_game():
    """Plays the game"""
    computer_choice = generate_computer_choice()
    user_choice = get_user_choice()
    print(f"Computer's choice: {computer_choice}")
    print(f"User's choice: {user_choice}")
    print(get_game_result(computer_choice, user_choice))


play_game()
````
